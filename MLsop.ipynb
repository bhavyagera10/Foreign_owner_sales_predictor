```python
def clean_dataset(df):
    assert isinstance(df, pd.DataFrame), "df needs to be a pd.DataFrame"
    df.dropna(inplace=True)
    indices_to_keep = ~df.isin([np.nan, np.inf, -np.inf]).any(1)
    return df[indices_to_keep].astype(np.float64)
```

```python
#Helper function to perform EDA on Numerical variables.
def analyze_numeric(data, feature_var, color_plot):
    #plt.figure()
    #print(data[feature_var].describe())
    print("Correlation between variables")
    print(data[[feature_var,'net sales']].corr())   #Print the correlation value.
    fig, axs = plt.subplots(figsize= (15,8),ncols=2) #creating figure with multiple axes.
    plt.subplots_adjust(right=1)
    sns.boxplot(data[feature_var], color = color_plot ,ax=axs[0]) #explains the distribution of the variable
    axs[0].set_title('Distribution of '+ feature_var)
    sns.regplot(x=feature_var, y='net sales', color = color_plot,  data=data, ax= axs[1]) #regression plot for label vs feature variable
    axs[1].set_title( feature_var+ ' Vs Net Sales')
```

```python
analyze_numeric(dataset, 'inventory', 'green')
```

```python
analyze_numeric(dataset, '% curr assets', 'blue')
```

```python
analyze_numeric(dataset, 'net profit', 'purple')
```

```python
analyze_numeric(dataset, 'emp costs', 'pink')
```
